Assignment 1 - Working with Numpy

      1. Create a 1D NumPy array with integers from 1 to  20. Perform the following
operations:

   a. Calculate the sum, mean, median, and standard deviation of the elements in the
      array.

   b. Find the indices of elements greater than 10

```python
# NumPy: basic array and statistics
import numpy as np

# 1D array 1..20
arr = np.arange(1, 21)   # [1,2,...,20]
print("Array:", arr)

# (a) sum, mean, median, std
arr_sum = arr.sum()
arr_mean = arr.mean()
arr_median = np.median(arr)
arr_std = arr.std(ddof=0)   # population std

print("\nStatistics:")
print("Sum:", arr_sum)
print("Mean:", arr_mean)
print("Median:", arr_median)
print("Std (population):", arr_std)

# (b) indices where element > 10
indices_gt_10 = np.where(arr > 10)[0]  # zero-based indices
print("\nIndices where arr > 10 (zero-based):", indices_gt_10)
print("Values > 10:", arr[arr > 10])

# (c) reshape to 4x5 matrix     
```

2. Create a 2D NumPy array of shape 4 X 4 with numbers ranging from 1 to 16.

     a. Print the array. 

     b. Find the transpose of the array. 
  
     c. Calculate the row-wise and column-wise sums of the array.

```python

# 2D array 4x4
arr2d = np.arange(1, 17).reshape(4, 4)
print("2D array:\n", arr2d)

# transpose
print("\nTranspose:\n", arr2d.T)

# row-wise sums and column-wise sums
row_sums = arr2d.sum(axis=1)
col_sums = arr2d.sum(axis=0)
print("\nRow-wise sums:", row_sums)
print("Column-wise sums:", col_sums)    
```

3. Create two 3 X 3 arrays filled with random integers between 1 and 20.

     a. Perform element-wise addition, subtraction, and multiplication. 
     
     b. Compute the dot product of the two arrays.

```python
# reproducible random 3x3 matrices
np.random.seed(0)
A = np.random.randint(1, 21, size=(3,3))
B = np.random.randint(1, 21, size=(3,3))

print("A:\n", A)
print("\nB:\n", B)

# a) element-wise add, sub, multiply
print("\nA + B:\n", A + B)
print("\nA - B:\n", A - B)
print("\nA * B (element-wise):\n", A * B)

# b) dot product (matrix multiplication)
print("\nA dot B:\n", A.dot(B))
# 1D array 1..20
arr = np.arange(1, 21)   # [1,2,...,20]
print("Array:", arr)        
```

4. Reshape a 1D array of size 12 into a 3 X 4 2D array and slice the first two rows
   and last two columns.

```python
arr12 = np.arange(1, 13)           # 1..12
arr3x4 = arr12.reshape(3, 4)       # shape (3,4)
print("3x4 array:\n", arr3x4)

# slice: first two rows (0:2), last two columns (-2:)
slice_result = arr3x4[:2, -2:]
print("\nSliced (first 2 rows, last 2 columns):\n", slice_result)
```

Assignment 2 - Working with Pandas

1. Create a DataFrame with the following data:

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}

df = pd.DataFrame(data)
df.head()   
```

 a. Print the first five rows of the DataFrame. 

 b. Get the summary statistics of the 'Age' and 'Salary' columns. 
 
 c. Calculate the average salary of employees in the 'HR' department.

```python
# (a) first five rows
print("First five rows:")
display(df.head())

# (b) summary stats for Age and Salary
print("\nSummary of Age and Salary:")
display(df[['Age','Salary']].describe())

# (c) average salary of HR department
avg_hr = df[df['Department'] == 'HR']['Salary'].mean()
print("\nAverage salary in HR:", avg_hr)
```

2. Add a new column, 'Bonus', which is 10% of the salary

```python
df['Bonus'] = df['Salary'] * 0.10
display(df)
```

3. Filter the DataFrame to show employees aged between 25 and 30.

```python
filtered = df[(df['Age'] >= 25) & (df['Age'] <= 30)]
display(filtered)
```

4. Group the data by 'Department' and calculate the average salary for each
   department.

```python
grouped = df.groupby('Department')['Salary'].mean().reset_index(name='Avg_Salary')
display(grouped)
```

5. Sort the DataFrame by 'Salary' in ascending order and save the result to a new CSV
file.

```python
sorted_df = df.sort_values(by='Salary', ascending=True).reset_index(drop=True)
display(sorted_df)

# Save
csv_path = "/mnt/data/pandas_assignment_sorted.csv"   # in Colab use drive path if needed
sorted_df.to_csv(csv_path, index=False)
print("Saved sorted CSV to:", csv_path)
```

Assignment 3 - Working with Matplotlib

1. Create a simple line plot for the following data:

      a. Plot the data. 

      b. Customize the plot by adding a title, axis labels, and a grid.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

plt.figure()
plt.plot(x, y)
plt.title("Line Plot: x vs y")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

2. Create a bar graph to represent the marks scored by students in a subject:

      a. Plot the data as a bar graph. 
      
      b. Customize the colors and add a title.

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

plt.figure()
plt.bar(students, marks)
plt.title("Marks by Student")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()
```

3. Create a pie chart to represent the percentage distribution of a company’s revenue
from different regions:

    a. Create a pie chart with the region names as labels. 
    
    b. Highlight the region with the highest revenue.

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

# find index of max to explode
explode = [0.1 if r == max(revenue) else 0 for r in revenue]

plt.figure()
plt.pie(revenue, labels=regions, autopct='%1.1f%%', startangle=90, explode=explode)
plt.title("Revenue Distribution by Region")
plt.axis('equal')
plt.show()
```

4. Generate a histogram to show the frequency distribution of randomly generated
integers between 1 and 100 (sample size = 1000).

```python
import numpy as np
np.random.seed(0)
data = np.random.randint(1, 101, size=1000)

plt.figure()
plt.hist(data, bins=20)
plt.title("Histogram of Random Integers (1-100), n=1000")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
