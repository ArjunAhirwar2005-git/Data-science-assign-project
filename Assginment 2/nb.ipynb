Q. 1.Basics of Probability
   >A. Tossing a coin 10,000 times and calculating the experimental probability of heads
      and tails.

```python
import random
from collections import Counter

# (a) Tossing a coin 10,000 times
N = 10000
outcomes = [random.choice(['H', 'T']) for _ in range(N)]
counts = Counter(outcomes)

p_heads = counts['H'] / N
p_tails = counts['T'] / N

print("=== Coin Toss Simulation ===")
print("Total tosses:", N)
print("Heads:", counts['H'], "Tails:", counts['T'])
print(f"P(Heads) = {p_heads:.4f}")
print(f"P(Tails) = {p_tails:.4f}")              
print()         
# --- a/vscode-notebook-cell:Untitled-1.ipynb?jupyter-notebook#W4sdW50aXRsZWQ%3D    
# +++ b/vscode-notebook-cell:Untitled-1.ipynb?jupyter-notebook#W4sdW50aXRsZWQ%3D


# (b) Rolling a six-sided die 10,000 times
N = 10000
rolls = [random.randint(1, 6) for _ in range(N)]
counts = Counter(rolls)

for face in range(1, 7):
    print(f"Face {face}: {counts[face]} times, Probability: {counts[face]/N:.4f}")  
print("=== Die Roll Simulation ===")
print("Total rolls:", N)
print()
```

Q.2. Write a function to estimate the probability of getting at least one "6" in 10 rolls
     of a fair die

     Steps 
   a. Simulate rolling a die 10 times using a loop. 
 
   b. Track trials where at least one "6" occurs.   

   c. Calculate the proportion of successful trials.

```python
def estimate_at_least_one_six(trials=20000, rolls=10):
    success = 0
    for _ in range(trials):
        if any(random.randint(1,6)==6 for _ in range(rolls)):
            success += 1
    return success / trials

estimated = estimate_at_least_one_six()
theoretical = 1 - (5/6)**10
print("=== Probability of At Least One 6 in 10 Rolls ===")
print(f"Experimental  = {estimated:.5f}")
print(f"Theoretical   = {theoretical:.5f}")

# --- a/vscode-notebook-cell:Untitled-1.ipynb?jupyter-notebook#W4sdW50aXRsZWQ%3D

# +++ b/vscode-notebook-cell:Untitled-1.ipynb?jupyter-notebook#W4sdW50aXRsZWQ%3D
# (e) Simulating a custom discrete distribution         
outcomes = [1, 2, 3]
probabilities = [0.25, 0.35, 0.4]
N_samples = 10000
samples = random.choices(outcomes, weights=probabilities, k=N_samples)      
counts = Counter(samples)
mean = sum(x * counts[x] for x in outcomes) / N_samples             
variance = sum(((x - mean) ** 2) * counts[x] for x in outcomes) / N_samples
std_dev = variance ** 0.5
print("=== Custom Discrete Distribution Simulation ===")
print("Total samples:", N_samples)  

for outcome in outcomes:

    print(f"Outcome {outcome}: {counts[outcome]} times, Empirical P: {counts[outcome]/N_samples:.4f}, Theoretical P: {probabilities[outcome-1]:.4f}")
print(f"Empirical Mean: {mean:.4f}")            
print(f"Empirical Standard Deviation: {std_dev:.4f}")   
print(f"Empirical Variance: {variance:.4f}")
# --- a/vscode-notebook-cell:Untitled-1.ipynb?jupyter-notebook#W4sdW50aXRsZWQ%3D
# +++ b/vscode-notebook-cell:Untitled-1.ipynb?jupyter-notebook#W4sdW50aXRsZWQ%3D
# (f) Simulating an exponential distribution
import numpy as np      
import matplotlib.pyplot as plt
mean_exp = 5
N_samples = 2000        
samples = np.random.exponential(scale=mean_exp, size=N_samples)
# Histogram
plt.hist(samples, bins=30, density=True, alpha=0.6, color='
blue', label='Histogram')
# PDF Overlay               
x = np.linspace(0, max(samples), 1000)
pdf = (1/mean_exp) * np.exp(-x/mean_exp)            
plt.plot(x, pdf, 'r-', lw=2, label='PDF Overlay')
plt.title('Exponential Distribution Simulation')
plt.xlabel('Value') 
plt.ylabel('Density')
plt.legend()
plt.show()          
# --- a/vscode-notebook-cell:Untitled-1.ipynb?jupyter-notebook#W4sdW50aXRsZWQ%3D    
```

3.  A bag contains 5 red, 7 green, and 8 blue balls. A ball is drawn randomly, its color
    noted, and it is put back into the bag. If this process is repeated 1000 times, write a
    Python program to estimate:

    a. The probability of drawing a red ball given that the previous ball was blue. 

    b. Verify Bayes' theorem with the simulation results.

```python
colors = ['red']*5 + ['green']*7 + ['blue']*8
draws = [random.choice(colors) for _ in range(1000)]

count_prev_blue = 0
count_red_given_blue = 0

for i in range(1, len(draws)):
    if draws[i-1] == 'blue':
        count_prev_blue += 1
        if draws[i] == 'red':
            count_red_given_blue += 1

empirical = count_red_given_blue / count_prev_blue
print("=== Conditional Probability Simulation ===")
print(f"P(current=red | previous=blue) = {empirical:.4f}")
print("Theoretical (independent draws) = 0.25")




        
```

4. Generate a sample of size 1000 from a discrete random variable with the following
distribution: 
 - P(X=1) = 0.25 
 - P(X=2) = 0.35 
 - P(X=3) = 0.4 
 Compute the empirical mean, variance, and standard deviation of the sample.

 Steps 

     a. Use numpy.random.choice() to generate the sample. 
  
     b. Use numpy methods to calculate mean, variance, and standard deviation.

```python
import numpy as np

values = [1, 2, 3]
prob = [0.25, 0.35, 0.40]

sample = np.random.choice(values, size=1000, p=prob)
mean_emp = sample.mean()
var_emp = sample.var()
std_emp = sample.std()

print("=== Discrete Random Variable ===")
print(f"Empirical mean = {mean_emp:.4f}")
print(f"Empirical variance = {var_emp:.4f}")
print(f"Empirical std dev = {std_emp:.4f}")

            
```

5. Simulate 2000 random samples from an exponential distribution with a mean of 5.
   Visualize the distribution using:

    a. A histogram. 

    b. A probability density function (PDF) overlay.

```python
import matplotlib.pyplot as plt

np.random.seed(0)
scale = 5
data = np.random.exponential(scale=scale, size=2000)

# Histogram
plt.figure()
plt.hist(data, bins=40, density=True)
plt.title("Exponential Distribution (mean=5)")
plt.xlabel("Value"); plt.ylabel("Density")
plt.show()

# Overlay PDF
x = np.linspace(0, max(data), 200)
pdf = (1/scale)*np.exp(-x/scale)
plt.figure()
plt.hist(data, bins=40, density=True)
plt.plot(x, pdf)
plt.title("Histogram with PDF Overlay")
plt.xlabel("Value"); plt.ylabel("Density")
plt.show()

```

6. Simulate the Central Limit Theorem by following these steps

    a. Generate 10,000 random numbers from a uniform distribution.

    b. Draw 1000 samples of size n = 30. 
 
    c. Calculate and visualize the distribution of sample means.

```python
np.random.seed(1)

data = np.random.uniform(0,1,10000)

# Plot original uniform distribution
plt.figure()
plt.hist(data, bins=40, density=True)
plt.title("Uniform(0,1) Distribution")
plt.xlabel("Value"); plt.ylabel("Density")
plt.show()

# Draw 1000 samples of size 30 → sample means
sample_means = [np.mean(np.random.choice(data, 30, replace=True))
                for _ in range(1000)]

plt.figure()
plt.hist(sample_means, bins=40, density=True)
plt.title("Distribution of Sample Means (n=30)")
plt.xlabel("Sample Mean"); plt.ylabel("Density")
plt.show()

print(f"Empirical mean of sample means: {np.mean(sample_means):.4f}")
print(f"Empirical std dev of sample means: {np.std(sample_means):.4f}")
```
